# 直接對x計算相對應的函數值，搭配plt.plot()就能畫圖了

```python
x = np.linspace(0,10,100)
```

```python
x
```

```python
plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```
