## [pandas學習資源](http://github.com/justmarkham/pandas-videos)
#### 用Groupby找到美國最常目擊UFO的州

```python
df = pd.read_csv("http://bit.ly/uforeports")
```

```python
df.head()
```

```python
df_state = df.groupby("State").count()
```

```python
df_state
```

```python
df_state.sort_values(by="Time",ascending=False)
```

```python
df_state
```

### sort有排，原先資料又沒有排

```python
df_state.sort_values(by="Time",ascending=False,inplace=True)
```

```python
df_state.head()
```

```python
df_state[:10].Time.plot(kind="bar")
```

## Loc定位法

```python
df = pd.DataFrame(np.random.randn(5,3),
                  index=list(range(1,6)),
                  columns=list("ABC"))
```

```python
df
```

```python
df[df.B>0]
```

```python
df[df.B>0]["C"]=0
```

```python
df
```

```python
df.loc[2:3,"B":"C"]
```

```python
df.loc[2,"B"]
```

```python
df.loc[2,"B"] = 1
```
